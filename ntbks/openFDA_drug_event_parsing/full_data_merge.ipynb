```python
import glob
import numpy as np
import pandas as pd
import pickle

data_dir = "../../data/openFDA_drug_event/"
er_dir = data_dir+'er_tables_memory_efficient/'
```

```python
primarykey='safetyreportid'
```

```python
patients = pd.read_csv(er_dir+'patient.csv.gz',
                       compression='gzip',
                       dtype={
                           'safetyreportid' : 'str',
                           'patient_custom_master_age' : 'float'
                       })
```

```python
report = (pd.read_csv(er_dir+'report.csv.gz',
                      compression='gzip',
                     dtype={
                         'safetyreportid' : 'str'
                     }))
report.head()
```

```python
patients_merge = patients.merge(report, on='safetyreportid', how='inner')
patients_merge.head()
```

```python
report_serious = pd.read_csv(er_dir+'report_serious.csv.gz',compression='gzip')
report_serious.head()
```

```python
patients_merge = patients_merge.merge(report_serious, on='safetyreportid', how='inner')
patients_merge.head()
```

```python
reporter = pd.read_csv(er_dir+'reporter.csv.gz',compression='gzip')
reporter.head()
```

```python
patients_merge = patients_merge.merge(reporter, on='safetyreportid', how='inner')
patients_merge.head()
```

```python
patients_merge.info()
```

```python
drug_characteristics_df = pd.read_csv('../../data/openFDA_drug_event/er_tables_memory_efficient/drugcharacteristics.csv.gz',compression='gzip')
```

```python
reaction_df = pd.read_csv("../../data/openFDA_drug_event/er_tables_memory_efficient/reactions.csv.gz", compression='gzip', low_memory=False)
```

```python
patients_merge = patients_merge.merge(drug_characteristics_df, on='safetyreportid', how='inner')
```

```python
patients_merge = patients_merge.merge(reaction_df, on='safetyreportid', how='inner')
```

```python
patients_merge.info()
```

```python
patients_merge.to_csv('../../data/patients_drug_reaction_report_full.csv', index=False)
```

```python
sampled_data = patients_merge.sample(n=1000, random_state=42)

print("sampling 1,000 row:")
print("\nnumber of row and column:", sampled_data.shape)
print("\ncolumns:", sampled_data.columns.tolist())
display(sampled_data.head(10))


sampled_data.to_csv('../../data/sampled_full_data.csv', index=False)
sampled_data.to_excel('../../data/sampled_full_data.xlsx', index=False)
```
