```python

# สรุปการค้นหา comma ในทุกคอลัมน์
print("สรุปการค้นหา comma ในแต่ละคอลัมน์:")
print("="*50)

for col in clean_drug_v2.columns:
    if clean_drug_v2[col].dtype == 'object':  # ตรวจสอบเฉพาะคอลัมน์ที่เป็น string
        comma_count = clean_drug_v2[col].str.contains(',', na=False).sum()
        print(f"{col}: {comma_count:,} แถว")

print("\n" + "="*50)
print("แถวที่มี comma ในคอลัมน์ใด ๆ:")
# สร้าง mask สำหรับแถวที่มี comma ในคอลัมน์ใด ๆ
has_comma_mask = False
for col in clean_drug_v2.columns:
    if clean_drug_v2[col].dtype == 'object':
        has_comma_mask |= clean_drug_v2[col].str.contains(',', na=False)

rows_with_any_comma = clean_drug_v2[has_comma_mask]
print(f"จำนวนแถวที่มี comma ในคอลัมน์ใด ๆ: {len(rows_with_any_comma):,}")
```

```python

# แสดงตัวอย่างแถวที่มี comma พร้อมรายละเอียด
print("ตัวอย่างแถวที่มี comma (แสดง 20 แถวแรก):")
print("="*80)
display(rows_with_any_comma.head(20))

print("\n" + "="*80)
print("ค้นหาเฉพาะแถวที่มี comma ในชื่อยา (medicinal_product_clean):")
if len(rows_with_comma_clean) > 0:
    print(f"พบ {len(rows_with_comma_clean)} แถว")
    print("\nรายการชื่อยาที่มี comma:")
    unique_drugs_with_comma = rows_with_comma_clean['medicinal_product_clean'].unique()
    for i, drug in enumerate(unique_drugs_with_comma[:30], 1):  # แสดง 30 รายการแรก
        print(f"{i:2d}. {drug}")
    
    if len(unique_drugs_with_comma) > 30:
        print(f"... และอีก {len(unique_drugs_with_comma) - 30} รายการ")
else:
    print("ไม่พบแถวที่มี comma ในคอลัมน์ medicinal_product_clean")
```
